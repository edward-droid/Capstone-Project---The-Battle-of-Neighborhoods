# Restaurant Expansion to Toronto

## Data
Most of the datasets have been obtained from the City of Toronto website Open Data Portal: https://www.toronto.ca/city-government/data-research-maps/open-data/ 

The following datasets have been obtained and used from the City of Toronto Open Data Portal:
#### 1) City Wards Data: 
This dataset contains the name and corresponding ward number for all 25 administrative wards of the City of Toronto. The data from this dataset will be used to create a dataframe with the Ward Name and Ward #.

#### 2) 2018-ward-profiles-25-ward-model-geographic-areas: 
This dataset contains the land area size in square kilometers of each Ward #. The data from this dataset will be used to determine the size of radius for each ward when mapping each ward's circle marker on Folium maps. For example, if a ward's land area is 48.38 square km, the radius of the ward would be assumed to be the square root of the Area of the Ward divided by the value of Pi:

r = Square Root of (Area of Ward / pi)

r = Square Root of (48.38 / 3.14)

r = 3.92km

This radius calculation can be used to draw circular markers on the map of Toronto to visualize each ward's location and size. In reality, the shape of each ward is not circular, but the actual shape is not relevant to the analysis of data for this project.

Land Area data will also be used to calculate population density calculations for each ward (Population of Ward / Land Area of Ward) and the density of food service businesses in each ward (# of Food Service Businesses in Ward / Land Area of Ward).   

#### 3) Food Inspection Dataset (City of Toronto): 
Every eating and drinking establishment in the City of Toronto receives a minimum of 1 to 3 inspections each year. Hence, this dataset contains the names, latitude, and longitude of every food establishment in Toronto. As a result, this dataset will be used to count how many food service establishments (potential competitors to ABCD Brands) there are in each ward of Toronto. This data can provide a picture of the supply of food service establishments (# of food service establishments in the ward) compared to the demand for food service establishments (# of people residing in the ward) in each ward. Ideally, it would be best to set up a restaurant in a neighborhood where there is a shortage of restaurants to serve the people in the area. Therefore, this data will help determine which wards are potential candidates based on supply of food service businesses and demand for food services business to set up ABCD's restaurant.

#### 4) 2018-ward-profiles-2011-2016-census-25-ward-model-data: 
This dataset contains demographic, social and economic information for each ward in Toronto. The information is obtained from 2016 and 2011 Census data for the 25 wards of Toronto. The types of data that will be used from this dataset are:
    
Total Population: this data will be used to understand the absolute number of people and population density of each ward. Having a restaurant in high density neighborhood is usually better than a low density neighborhood and high total population is better than low population in ward.

Population Segmented by Age Group: this data will be used to understand the age demographic of each ward. In general, age segments that are of working age (24 to 65) are preferred since this age demographic has more money to spend

Population Based on Country Ethnic Origin: this data will be used to understand what specific country category of food may be in high demand. For example, if the neighborhood has many people from Brazil, restaurants serving Brazilian food may be suitable for the ward.

Population Segmented by Income: this data will be used to understand what price point restaurant could do well in the neighborhood. For example, if the neighborhood has many residents with a high level of income, a restaurant serving high priced food may do well.
    
Difference between 2011 and 2016 population: this data will help determine which wards are growing the most. Ideally, we would want to set up a restaurant in a ward with a growing population.

#### 5) Ward Visible Minorities Data: 
This dataset provides data on the racial ethnicity of population in the ward. For example, rather than separating Colombian and Venezuelans, all the people for Latin America are grouped as Latin American in this dataset.  

#### 6) Google Maps: 
Each ward's geographic profile map is provided on the City of Toronto website such as in the sample pdf displayed below. Since the geographic center of an irregular shaped object such as a city ward cannot be defined absolutely, it was necessary to look at the profile map of each ward on the City of Toronto website and estimate an arbitrary center point using road intersection or major landmark around the center of each ward. For example, in the image below, the geographic center would be around the area of Dundas Street a bit east of Sherbourne Street. By finding and clicking the same intersection or landmark of the center point of the ward on Google Maps, the corresponding Latitude and Longitude of the location is displayed on Google Maps. This process was repeated for each of the 25 wards to determine an arbitrary center point of each ward. The center point latitude and longitude of each ward will be used to map each ward on Folium map of Toronto. 

Additionally, this data on center point of each ward will be used to determine how many food service establishments each ward has. Knowing the latitude and longitude coordinates for every food service establishment from the dataset in 3) above, each establishment is categorized as belonging to the ward whose center point that it is closest to. Assigning each restaurant to a ward based on distance to ward center point instead of the actual ward it belongs to is arguably the best way to categorize each establishment's ward category since each restaurant is affected by the demographics of the ward that it is geographically closest to even though it may administratively belong to a different ward. Ideally, we would like to set up a restaurant in a ward where the number of food service establishments are very few relative to the population and population density of the ward. Knowing how many food service establishments are in each ward will help us achieve this.



#### 7) Foursquare API: 
Search will be conducted to find what the most common type of restaurants and other type of venues are for each ward. This information will provide a better picture of what type of businesses are existing in each ward. Restaurants can benefit from what type of other venues there are in the neighborhood. For example, a restaurant can have a large source of customers if there are other retail shops in the neighborhood. In addition, many times restaurant customers will first visit an area to go shopping or do other activities before eating so the existence of other venues that draw people can be a source of customers. The ideal neighborhood will have other venues to draw in potential restaurant customers from other neighborhoods so that the restaurant will not have to depend on only customers living in the same ward that the restaurant is located in. Foursquare location data will help find neighborhoods with this characteristic.
